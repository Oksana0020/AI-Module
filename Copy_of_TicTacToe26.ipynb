```python
class CurrentBoard():

    def __init__(self, setup_of_board= "         "):
      self.board = setup_of_board
      self.state = self.state_of_play()

    def display_board(self, game_display = False):
      c = ""

      if game_display:
        index_display = 0
        for index in range(len(self.board)):
          if self.board[index] == " ":
              c += str(index_display)
              index_display += 1
          else:
              c += self.board[index]
      else:
        c = self.board




      print(c[0] + '|' + c[1] + '|' + c[2])
      print('-+-+-')
      print(c[3] + '|' + c[4] + '|' + c[5])
      print('-+-+-')
      print(c[6] + '|' + c[7] + '|' + c[8])

    def all_possible_moves(self, forPlayer = "X"):
      all_moves = []
      for i in range(len(self.board)):
        if self.board[i] == " ":
          string_for_board = self.board[:i] + forPlayer + self.board[i+1:]
          all_moves.append(CurrentBoard(setup_of_board=string_for_board))
      return all_moves

    def other_player(self, forPlayer):
      if forPlayer == "X":
        return "O"
      return "X"

    def Eq3(self, i1,i2,i3):
      a = self.board[i1]
      b = self.board[i2]
      c = self.board[i3]
      return (a==b) and (b==c) and (a!=" ")

    def state_of_play(self):
    # "X" if X wins
    # "O" if O wins
    # "D" if draw
    # "U" if unfinished

    #012
    #345
    #678

      if self.Eq3(0,1,2) or self.Eq3(0,3,6) or self.Eq3(0,4,8):
        return self.board[0]
      if self.Eq3(1,4,7) or self.Eq3(3,4,5) or self.Eq3(2,4,6):
        return self.board[4]
      if self.Eq3(2,5,8) or self.Eq3(6,7,8):
        return self.board[8]
      if " " in self.board:
        return "U"
      return "D"
```

```python
class SearchTreeNode():
  def __init__(self, current_board, forPlayer,ply =0):
    self.board = current_board
    self.ply_depth = ply
    self.value_is_assigned = False
    self.children = []
    if current_board.state == "U":
     for move in current_board.all_possible_moves(forPlayer):
      self.children.append(SearchTreeNode(move, self.board.other_player(forPlayer), self.ply_depth +1))

    else: # Game is Over so this node is a leaf!!
      self.value_is_assigned = True
      if self.board.state == "D":
        self.value = 0
      else:
        if self.ply_depth % 2 == 0:
          self.value = -1
        else:
          self.value = 1


  def min_max_value(self):
    if self.value_is_assigned:
       return self.value
      # value is not assigned
    self.children = sorted(self.children,key=lambda x:x.min_max_value())

    self.value_is_assigned = True

    if self.ply_depth % 2 == 0:
      self.value = self.children[-1].value
    else:
      self.value = self.children[0].value

    return self.value
```

```python
def replace_space(board_string,index, charXorO):
  # replaces indexedTH  space in board_string with the character charXorO
  space_ind = 0
  board_index = 0
  for c in board_string:
    print(c)
    if c == " ":

      if space_ind == index:
        return board_string[:board_index] + charXorO + board_string[board_index+1:]
      space_ind += 1
    board_index += 1
  return board_string
```

```python
def play_game():
  response = input(" Do you want to play as X or O?"  )
  player_playing_as = response
  AI_playing_as = CurrentBoard().other_player(player_playing_as)
  response = input(" Do you want to play first Y/N?"  )
  player_playing_first =(response == "Y" or response == "y")
  player_turn = player_playing_first
  cb = CurrentBoard()

  while True:
    if player_turn:
      # display current board for player
      cb.display_board(game_display=True)
      player_move = input("Make your choice ")
      player_move = int(player_move)
      boards_string = replace_space(cb.board, player_move, player_playing_as)
      cb = CurrentBoard(boards_string)

    else:
      #   AI move
      print(" AI is making the move for the following board, playing as " +AI_playing_as)
      cb.display_board()
      print("Constructing the search tree......")
      st = SearchTreeNode(cb, AI_playing_as)
      print("Applying the min Max algorithm")
      st.min_max_value()
      print(" Choosing the best move")
      cb = st.children[-1].board

    # Check for end game
    if cb.state != "U":
      cb.display_board()
    if cb.state == "D":
      print("Game is a draw, I can't believe it, you got lucky")
      return
    if cb.state == AI_playing_as:
      print(" I win too easy!!!!")
      return
    if cb.state == player_playing_as:
      print("This is just not possible")
      return
    # Toggle AI and Player move
    player_turn = not player_turn
```
