```python
class CurrentBoard:

    def __init__(self, setup_of_board="         "):
        # Board is a string of 9 characters: "X", "O", or " "
        self.board = setup_of_board

    def display_board(self):
        c = self.board
        print(c[0] + "|" + c[1] + "|" + c[2])
        print("_+_+_")
        print(c[3] + "|" + c[4] + "|" + c[5])
        print("_+_+_")
        print(c[6] + "|" + c[7] + "|" + c[8])
        print()

    def all_possible_moves(self, forPlayer="X"):
        all_moves = []
        for i in range(len(self.board)):
            if self.board[i] == " ":
                string_for_board = self.board[:i] + forPlayer + self.board[i+1:]
                all_moves.append(i)
        return all_moves
```

```python
class SearchTreeNode:
    def __init__(self, current_board, forPlayer):
        self.board = current_board
        self.children = []

        if current_board.state() == "U":
            for move in current_board.all_possible_moves(forPlayer):
                self.children.append(
                    SearchTreeNode(move, self.other_player(forPlayer))
                )

        else:   # game is over so this node is leaf
            if self.board.state == "X":
                self.value = 1
            elif self.board.state == "O":
                self.value = -1
            elif self.board.state == "D":
                self.value = 0

        # moves are indices, so convert each move into a NEW board string
        for move in current_board.all_possible_moves(forPlayer):
            new_board_string = (
                current_board.board[:move] +
                forPlayer +
                current_board.board[move+1:]
            )
            child_board = CurrentBoard(new_board_string)

            self.children.append(
                SearchTreeNode(child_board, self.other_player(forPlayer))
            )

    def other_player(self, forPlayer):
        if forPlayer == "X":
            return "O"
        else:
            return "X"

    def Eq3(self, i1, i2, i3):
        a = self.board[i1]
        b = self.board[i2]
        c = self.board[i3]
        return (a == b) and (b == c) and (a != " ")

    def state_of_play(self):
        # 'X' IF X WINS
        # 'O' IF O WINS
        # 'D' FOR DRAW
        # 'U' IS UNFINISHED

        # 012
        # 345
        # 678

        if self.Eq3(0, 1, 2) or self.Eq3(0, 3, 6) or self.Eq3(0, 4, 8):
            return self.board[0]

        if self.Eq3(1, 4, 7) or self.Eq3(3, 4, 5) or self.Eq3(2, 4, 6):
            return self.board[4]

        if self.Eq3(2, 5, 8) or self.Eq3(6, 7, 8):
            return self.board[8]

        if " " in self.board:
            return "U"

        return "D"
```
